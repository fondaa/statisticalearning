```r
# D
for (v in colnames(Boston)) {
    if (v != 'crim') {
        fit_formula <- as.formula(paste0('crim ~ ', v, '+ I(', v, '^2)', '+ I(', v, '^3)'))
        lr_tmp <- lm(fit_formula, data=Boston)
        print(summary(lr_tmp))
    }
}
```
